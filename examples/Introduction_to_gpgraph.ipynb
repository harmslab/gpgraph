```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from gpmap import GenotypePhenotypeMap
from gpgraph import GenotypePhenotypeGraph
from gpgraph.pyplot import draw_paths, flattened
from gpgraph.paths import forward_paths_prob, forward_paths
from gpgraph.models import strong_selection_weak_mutation
```

```python
# Import the data
gpm = GenotypePhenotypeMap.read_json('data/weinreich.json')
```

```python
# Parameters, the source and target genotypes
source="00000"
target="11111"

# Create GenotypePhenotypeGraph object from GenotypePhenotypeMap object
G = GenotypePhenotypeGraph(gpm)

# Add a transition model to the edges of the GenotypePhenotypeGraph object
G.add_model(strong_selection_weak_mutation)

# Calculate all the forward paths from source genotype to target genotype
# and then calculate their probability
forward_paths = forward_paths(G, source, target)
probability = forward_paths_prob(G, source, target)


fig, ax = plt.subplots(figsize=(4,4))
x = draw_paths(G,
               ax=ax,
               edge_scalar=9e6)

# Save figure as a pdf file
fig.savefig('generated_figures/example_figure.pdf', format='pdf', dpi=250)
```
